```python
# Import pytrends and other dependencies
import pandas as pd
from pytrends.request import TrendReq
from pathlib import Path
from sqlalchemy import create_engine
import time
import datetime
```

```python
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Maison Margiela','maison margiela jeans','MM6','MM6 jeans'], timeframe='today 12-m')
mm_df = pytrend.interest_by_region()
mm_df = mm_df[mm_df['Maison Margiela']> 0]
#mm_df
mm_df2 = pytrend.interest_over_time()
mm_df2
```

```python
# Export data to csv file
mm_df.to_csv("output_data/mm6.csv", index_label="geoName")
```

```python
# Related Queries
related_queries = pytrend.related_queries()
print(related_queries)
```
